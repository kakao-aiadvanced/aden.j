```python
def get_gpt4_mini_response(prompt):
    try:
        completion = client.chat.completions.create(
            model="gpt-4o-mini",  # 사용하려는 모델 지정
            messages=[
                # {"role": "system", "content": "you are"}
                {"role": "user", "content": prompt}
                ],
            max_tokens=150,  # 생성할 최대 토큰 수
            # temperature=1,  # 창의성 조절 (0~1 사이의 값)
            # n=1,  # 생성할 응답 수
            # stop=None  # 중지 토큰 설정
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {str(e)}"

# 예시 프롬프트
prompt = """Translate the following words to Korean:
1. book: 책
2. cheese: 치즈
3. water: 물
4. bread: 빵
5. apple: 사과

Translate the word "dog":"""

# GPT-4-mini 모델을 사용하여 응답을 가져옵니다.
response = get_gpt4_mini_response(prompt)
print(response)
```
